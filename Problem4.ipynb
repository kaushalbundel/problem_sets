# Problem

Find the largest palindrome made from the product of two 3-digit numbers.


# Solution

1. Creating a function to check if a number is pallindrome our not
2. Reiterate the function over the product of all three digit numbers
3. Find out the maximum number that is a pallindrome

```python
def is_pallindrome(number: int):
    num=str(number)
    if num[::-1] == num:
        return number
    else:
        return f"The number {number} is not a pallindrome"
```

```python
is_pallindrome(1001), is_pallindrome(102)
```
